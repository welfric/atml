```python
# ========================================
# Citations:
# - FedAvg: McMahan et al., AISTATS 2017
# - FedProx: Li et al., MLSys 2020
# - SCAFFOLD: Karimireddy et al., ICML 2020
# - FedGH: Zhang et al., IEEE SPL 2024
# - FedSAM: Qu et al., ICML 2022
# ========================================

# --------------------------------------------------------------
# 1. IMPORTS & SETUP
# --------------------------------------------------------------
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
import torchvision
import torchvision.transforms as transforms
import numpy as np
import matplotlib.pyplot as plt
from torch.utils.data import DataLoader, Subset
import copy
from collections import defaultdict
import pandas as pd

plt.rcParams.update({'font.size': 16, 'figure.figsize': (10, 7)})

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Using device: {device}')

# --------------------------------------------------------------
# 2. HYPER-PARAMETERS
# --------------------------------------------------------------
BATCH_SIZE   = 64
LEARNING_RATE = 0.01
LOCAL_EPOCHS  = 5
NUM_CLIENTS   = 10
COMM_ROUNDS   = 20
ALPHA         = 0.1          # Dirichlet concentration
FRAC          = 0.1        # fraction of the whole train set to use
MU            = 0.01         # FedProx
RHO           = 0.05         # FedSAM

# --------------------------------------------------------------
# 3. DATASET + TRANSFORMS
# --------------------------------------------------------------
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
])

full_trainset = torchvision.datasets.CIFAR10(root='./data', train=True,
                                            download=True, transform=transform)
testset = torchvision.datasets.CIFAR10(root='./data', train=False,
                                       download=True, transform=transform)
testloader = DataLoader(testset, batch_size=128, shuffle=False, num_workers=2)

# --------------------------------------------------------------
# 4. TAKE ONLY `frac` OF THE TRAINING DATA
# --------------------------------------------------------------
total_train = len(full_trainset)
keep = int(FRAC * total_train)
indices = np.random.permutation(total_train)[:keep]      # random subset
train_subset = Subset(full_trainset, indices)

print(f'Using {keep:,} / {total_train:,} training samples ({FRAC*100:.0f}% of CIFAR-10)')

# --------------------------------------------------------------
# 5. DIRICHLET PARTITION (now on the *subset*)
# --------------------------------------------------------------
def dirichlet_partition(dataset, num_clients, alpha, num_classes=10):
    """Return list of index-lists (one per client) from a *dataset*."""
    labels = np.array([dataset[i][1] for i in range(len(dataset))])
    client_indices = [[] for _ in range(num_clients)]

    for k in range(num_classes):
        idx_k = np.where(labels == k)[0]
        np.random.shuffle(idx_k)

        proportions = np.random.dirichlet(np.repeat(alpha, num_clients))
        
        split = (np.cumsum(proportions) * len(idx_k)).astype(int)[:-1]
        splits = np.split(idx_k, split)

        for c, split in enumerate(splits):
            client_indices[c].extend(split.tolist())

    for c in range(num_clients):
        np.random.shuffle(client_indices[c])

    return client_indices

# --------------------------------------------------------------
# 6. BUILD PER-CLIENT LOADERS
# --------------------------------------------------------------
client_idx_lists = dirichlet_partition(train_subset, NUM_CLIENTS, ALPHA)

client_loaders = []
for idxs in client_idx_lists:
    loader = DataLoader(Subset(train_subset, idxs),
                        batch_size=BATCH_SIZE,
                        shuffle=True,
                        drop_last=True,
                        num_workers=2)
    client_loaders.append(loader)

print(f'Partitioned into {NUM_CLIENTS} clients (total samples used: {sum(len(l.dataset) for l in client_loaders):,})')

class SimpleCNN(nn.Module):
    def __init__(self):
        super(SimpleCNN, self).__init__()
        self.conv1 = nn.Conv2d(3, 64, 5)
        self.pool = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(64, 64, 5)
        # Remove dropout1
        self.fc1 = nn.Linear(64 * 5 * 5, 384)
        # Remove dropout2
        self.fc2 = nn.Linear(384, 192)
        self.fc3 = nn.Linear(192, 10)
    
    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        # Remove dropout1(x)
        x = x.view(-1, 64 * 5 * 5)
        x = F.relu(self.fc1(x))
        # Remove dropout2(x)
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x
```

```python


# --------------------------------------------------------------
# EVALUATION FUNCTIONS (from provided)
# --------------------------------------------------------------
def evaluate_model(model, dataloader):
    model.eval()
    correct = total = 0
    criterion = nn.CrossEntropyLoss()
    loss_sum = 0.0
    with torch.no_grad():
        for x, y in dataloader:
            x, y = x.to(device), y.to(device)
            logits = model(x)
            loss_sum += criterion(logits, y).item()
            _, pred = logits.max(1)
            correct += pred.eq(y).sum().item()
            total += y.size(0)
    return 100.0 * correct / total, loss_sum / len(dataloader)

# --------------------------------------------------------------
# AGGREGATION AND UTILITY FUNCTIONS (from provided)
# --------------------------------------------------------------
def aggregate_models(global_model, client_states, weights):
    """Weighted average of state_dicts."""
    global_dict = global_model.state_dict()
    for k in global_dict:
        global_dict[k] = sum(w * client_states[i][k].float()
                             for i, w in enumerate(weights))
    global_model.load_state_dict(global_dict)
    return global_model

def flatten_params(state_dict):
    return torch.cat([p.view(-1) for p in state_dict.values()])

def unflatten_params(flat, ref):
    out = {}
    i = 0
    for k, p in ref.items():
        n = p.numel()
        out[k] = flat[i:i+n].view(p.shape)
        i += n
    return out

def compute_model_divergence(s1, s2):
    d = 0.0
    n = 0
    for k in s1:
        if 'weight' in k or 'bias' in k:
            diff = s1[k].float() - s2[k].float()
            d += (diff**2).sum().item()
            n += diff.numel()
    return np.sqrt(d / n) if n else 0.0

# --------------------------------------------------------------
# CLIENT TRAINING FUNCTIONS (from provided, including SCAFFOLD functional)
# --------------------------------------------------------------
def train_fedavg(model, loader, epochs, lr):
    model.train()
    opt = optim.SGD(model.parameters(), lr=lr, momentum=0.9, weight_decay=5e-4)
    # opt = optim.SGD(model.parameters(), lr=lr, momentum=0.5, weight_decay=1e-3)
    crit = nn.CrossEntropyLoss()
    loss_sum = correct = total = 0
    for _ in range(epochs):
        for x, y in loader:
            x, y = x.to(device), y.to(device)
            opt.zero_grad()
            loss = crit(model(x), y)
            loss.backward()
            opt.step()
            loss_sum += loss.item()
            _, pred = model(x).max(1)
            correct += pred.eq(y).sum().item()
            total += y.size(0)
    return model.state_dict(), loss_sum/(len(loader)*epochs), 100*correct/total

def train_fedprox(model, global_state, loader, epochs, lr, mu):
    model.train()
    opt = optim.SGD(model.parameters(), lr=lr, momentum=0.9, weight_decay=5e-4)
    crit = nn.CrossEntropyLoss()
    loss_sum = correct = total = 0
    for _ in range(epochs):
        for x, y in loader:
            x, y = x.to(device), y.to(device)
            opt.zero_grad()
            loss = crit(model(x), y)
            prox = sum(torch.sum((p - global_state[k].to(device))**2)
                       for k, p in model.named_parameters())
            (loss + 0.5*mu*prox).backward()
            opt.step()
            loss_sum += loss.item()
            _, pred = model(x).max(1)
            correct += pred.eq(y).sum().item()
            total += y.size(0)
    return model.state_dict(), loss_sum/(len(loader)*epochs), 100*correct/total

def train_fedsam(model, loader, epochs, lr, rho):
    model.train()
    base_optimizer = optim.SGD(model.parameters(), lr=lr, momentum=0.9, weight_decay=5e-4)
    criterion = nn.CrossEntropyLoss()
    
    loss_sum = correct = total = 0
    
    for _ in range(epochs):
        for x, y in loader:
            x, y = x.to(device), y.to(device)
            
            # First forward-backward pass (compute gradients)
            base_optimizer.zero_grad()
            loss = criterion(model(x), y)
            loss.backward()
            
            # Save original parameters
            old_params = [p.clone() for p in model.parameters()]
            
            # Compute gradient norm
            grad_norm = torch.sqrt(sum(p.grad.norm()**2 for p in model.parameters() if p.grad is not None))
            
            # Ascent step: perturb in gradient direction
            if grad_norm > 1e-12:
                with torch.no_grad():
                    for p in model.parameters():
                        if p.grad is not None:
                            p.add_(p.grad, alpha=rho / (grad_norm + 1e-12))
            
            # Second forward-backward pass at perturbed point
            base_optimizer.zero_grad()
            loss_pert = criterion(model(x), y)
            loss_pert.backward()
            
            # Restore original parameters before optimizer step
            with torch.no_grad():
                for p, p_old in zip(model.parameters(), old_params):
                    p.copy_(p_old)
            
            # Descent step with gradients from perturbed point
            base_optimizer.step()
            
            # Bookkeeping
            loss_sum += loss.item()
            _, pred = model(x).max(1)
            correct += pred.eq(y).sum().item()
            total += y.size(0)
    
    return model.state_dict(), loss_sum/(len(loader)*epochs), 100*correct/total

# --------------------------------------------------------------
# FEDGH (gradient harmonization) (from provided)
# --------------------------------------------------------------
def harmonize_gradients(updates_flat):
    """Pair-wise orthogonalisation of conflicting updates."""
    n = len(updates_flat)
    h = [u.clone() for u in updates_flat]
    for i in range(n):
        for j in range(i+1, n):
            dot = torch.dot(h[i], h[j])
            if dot < 0:
                ni2 = torch.sum(h[i]**2)
                nj2 = torch.sum(h[j]**2)
                if nj2 > 0 and ni2 > 0: # Add safety check
                    h[i] = h[i] - (dot / nj2) * h[j]
                    h[j] = h[j] - (dot / ni2) * h[i]
    return h

def train_client_scaffold(model, c_global, c_local, dataloader, epochs, lr):
    model.train()
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.SGD(model.parameters(), lr=lr)
    
    # Save initial model
    initial_params = [p.clone().detach() for p in model.parameters()]
    
    total_samples = 0
    correct = 0
    
    for epoch in range(epochs):
        for data, target in dataloader:
            data, target = data.to(device), target.to(device)
            optimizer.zero_grad()
            output = model(data)
            loss = criterion(output, target)
            loss.backward()
            
            # SCAFFOLD correction
            with torch.no_grad():
                for p, cg, cl in zip(model.parameters(), c_global, c_local):
                    p.grad.data += (cg - cl)
            
            optimizer.step()
            
            _, predicted = output.max(1)
            total_samples += target.size(0)
            correct += predicted.eq(target).sum().item()
    
    # Compute NEW control variate using Option II formula
    c_new_list = []
    total_steps = epochs * len(dataloader)
    
    with torch.no_grad():
        # CORRECTED: Include c_local (c_i) in the formula!
        for p_new, p_init, cg, cl in zip(model.parameters(), initial_params, c_global, c_local):
            # Option II: c_i^+ = c_i - c + (x - y) / (K * η)
            c_new = cl - cg + (p_init - p_new) / (total_steps * lr)
            c_new_list.append(c_new)
    
    # Compute delta for aggregation
    c_delta = [c_new - cl for c_new, cl in zip(c_new_list, c_local)]
    
    train_acc = 100. * correct / total_samples
    return c_delta, train_acc

# -------------------------------------------------
# MODIFIED FEDERATED LEARNING FUNCTION (integrated)
# -------------------------------------------------
def federated_learning(num_clients, num_rounds, local_epochs, lr,
                       alpha, method='fedavg',
                       mu=MU, rho=RHO, batch_size=BATCH_SIZE):
    print('\n' + '='*70)
    print(f'Training with {method.upper()} (α={alpha}')
    print('='*70)

    # ----- global structures -----
    global_model = SimpleCNN().to(device)

    # SCAFFOLD: Initialize server and local controls as lists of tensors
    if method == 'scaffold':
        c_global = [torch.zeros_like(p) for p in global_model.parameters()]
        c_locals = [[torch.zeros_like(p) for p in global_model.parameters()] for _ in range(num_clients)]

    # ----- metrics containers -----
    metrics = {
        'test_acc': [], 'test_loss': [], 'train_acc': [], 'train_loss': [],
        'divergence': [], 'grad_norm': [], 'cos_sim': [], 'upd_mag': []
    }

    for rnd in range(num_rounds):
        client_states = []
        client_weights = []
        client_train_acc = []
        client_train_loss = []
        divergences = []

        global_before = copy.deepcopy(global_model.state_dict())

        # ---------- client updates ----------
        if method == 'scaffold':
            # Use functional SCAFFOLD
            local_cs = []
            for cid in range(num_clients):
                # Initialize client model
                client_model = SimpleCNN().to(device)
                client_model.load_state_dict(copy.deepcopy(global_before))

                # Get client loader
                loader = client_loaders[cid]  # Assuming client_loaders is defined
                # SCAFFOLD_LR = 0.1
                # Train
                c_delta, train_acc = train_client_scaffold(
                    client_model, c_global, c_locals[cid],
                    loader, local_epochs, lr
                )

                state = client_model.state_dict()
                client_states.append(state)
                local_cs.append(c_delta)  # This is delta, not new c
                client_weights.append(len(loader.dataset))
                client_train_loss.append(0.0)  # Functional doesn't return loss, placeholder
                client_train_acc.append(train_acc)
                divergences.append(compute_model_divergence(state, global_before))
        else:
            # Other methods use functional approach
            for cid, loader in enumerate(client_loaders):
                model = SimpleCNN().to(device)
                model.load_state_dict(copy.deepcopy(global_before))

                if method == 'fedavg':
                    state, loss, acc = train_fedavg(model, loader, local_epochs, lr)
                elif method == 'fedprox':
                    state, loss, acc = train_fedprox(model, global_before, loader,
                                                    local_epochs, lr, mu)
                elif method == 'fedsam':
                    state, loss, acc = train_fedsam(model, loader, local_epochs, lr, rho)
                elif method == 'fedgh':
                    state, loss, acc = train_fedavg(model, loader, local_epochs, lr)
                else:
                    raise ValueError(f'Unknown method {method}')

                client_states.append(state)
                client_weights.append(len(loader.dataset))
                client_train_loss.append(loss)
                client_train_acc.append(acc)
                divergences.append(compute_model_divergence(state, global_before))

        # ---------- aggregation ----------
        total_samples = sum(client_weights)
        w_norm = [w/total_samples for w in client_weights]

        if method == 'fedgh':
            # FedGH gradient harmonization
            updates = [flatten_params(s) - flatten_params(global_before)
                       for s in client_states]
            harm_updates = harmonize_gradients(updates)
            avg_upd = sum(w*up for w, up in zip(w_norm, harm_updates))
            new_global_flat = flatten_params(global_before) + avg_upd
            new_global = unflatten_params(new_global_flat, global_before)
            global_model.load_state_dict(new_global)
        else:
            # Standard weighted averaging (FedAvg, FedProx, FedSAM, SCAFFOLD)
            global_model = aggregate_models(global_model, client_states, w_norm)

        # SCAFFOLD: Update server and local controls
        if method == 'scaffold':
            # Update c_global with average of deltas
            for param_idx in range(len(c_global)):
                delta_avg = sum(delta[param_idx] for delta in local_cs) / num_clients
                c_global[param_idx] += delta_avg

            # Update c_locals
            for cid in range(num_clients):
                for param_idx in range(len(c_locals[cid])):
                    c_locals[cid][param_idx] += local_cs[cid][param_idx]

        # ---------- evaluation ----------
        test_acc, test_loss = evaluate_model(global_model, testloader)
        metrics['test_acc'].append(test_acc)
        metrics['test_loss'].append(test_loss)
        metrics['train_acc'].append(np.mean(client_train_acc))
        metrics['train_loss'].append(np.mean(client_train_loss))
        metrics['divergence'].append(np.mean(divergences))

        # Gradient metrics
        upd_flat = [flatten_params(s) - flatten_params(global_before)
                    for s in client_states]
        norms = [u.norm().item() for u in upd_flat]
        metrics['grad_norm'].append(np.mean(norms))

        cos = []
        for i in range(num_clients):
            for j in range(i+1, num_clients):
                norm_i = upd_flat[i].norm()
                norm_j = upd_flat[j].norm()
                if norm_i > 1e-8 and norm_j > 1e-8:
                    c = torch.dot(upd_flat[i], upd_flat[j]) / (norm_i * norm_j)
                    cos.append(c.item())
        metrics['cos_sim'].append(np.mean(cos) if cos else 0.0)
        metrics['upd_mag'].append(np.mean([u.norm().item() for u in upd_flat]))

        if (rnd+1) % 5 == 0:
            print(f'R{rnd+1:02d}/{num_rounds} | '
                  f'TestAcc {test_acc:5.2f}% | TrainAcc {np.mean(client_train_acc):5.2f}% | '
                  f'Div {np.mean(divergences):.5f}')

    return metrics



##END NEW STUFF###

# --------------------------------------------------------------
# 11. TABLE PRINTING FUNCTIONS
# --------------------------------------------------------------
def print_summary_table(all_results, methods_config):
    """Print a comprehensive summary table."""
    print('\n' + '='*120)
    print('COMPREHENSIVE RESULTS SUMMARY TABLE')
    print('='*120)

    data = []
    label_map = {'fedavg': 'FedAvg', 'fedprox': 'FedProx', 'scaffold': 'SCAFFOLD',
                 'fedgh': 'FedGH', 'fedsam': 'FedSAM'}

    communication_costs = {
        'fedavg': 1.0, 'fedprox': 1.0, 'scaffold': 2.0, 'fedgh': 1.0, 'fedsam': 1.0
    }
    computation_costs = {
        'fedavg': 1.0, 'fedprox': 1.05, 'scaffold': 1.1, 'fedgh': 1.02, 'fedsam': 2.0
    }

    for method_name in methods_config.keys():
        final_test_acc = all_results[method_name]['test_acc'][-1]
        final_test_loss = all_results[method_name]['test_loss'][-1]
        final_train_acc = all_results[method_name]['train_acc'][-1]
        avg_divergence = np.mean(all_results[method_name]['divergence'])
        gen_gap = final_train_acc - final_test_acc

        data.append({
            'Method': label_map[method_name],
            'Test Acc (%)': f'{final_test_acc:.2f}',
            'Test Loss': f'{final_test_loss:.4f}',
            'Train Acc (%)': f'{final_train_acc:.2f}',
            'Gen Gap (%)': f'{gen_gap:.2f}',
            'Avg Divergence': f'{avg_divergence:.5f}',
            'Comm Cost': f'{communication_costs[method_name]:.1f}x',
            'Comp Cost': f'{computation_costs[method_name]:.1f}x'
        })

    df = pd.DataFrame(data)
    print(df.to_string(index=False))
    print('='*120)

def print_convergence_table(all_results, methods_config, threshold=60.0):
    """Print convergence analysis table."""
    print('\n' + '='*80)
    print(f'CONVERGENCE ANALYSIS (Threshold: {threshold}% Test Accuracy)')
    print('='*80)

    data = []
    label_map = {'fedavg': 'FedAvg', 'fedprox': 'FedProx', 'scaffold': 'SCAFFOLD',
                 'fedgh': 'FedGH', 'fedsam': 'FedSAM'}

    for method_name in methods_config.keys():
        test_accs = all_results[method_name]['test_acc']

        converged_round = None
        for i, acc in enumerate(test_accs):
            if acc >= threshold:
                converged_round = i + 1
                break

        if len(test_accs) >= 10:
            early_acc = np.mean(test_accs[:10])
            late_acc = np.mean(test_accs[-10:])
            improvement = late_acc - early_acc
        else:
            improvement = test_accs[-1] - test_accs[0]
            early_acc = test_accs[0]
            late_acc = test_accs[-1]

        data.append({
            'Method': label_map[method_name],
            'Converged at Round': converged_round if converged_round else 'No',
            'Early Acc (R1-10)': f'{early_acc:.2f}%',
            'Late Acc (R41-50)': f'{late_acc:.2f}%',
            'Improvement': f'{improvement:.2f}%'
        })

    df = pd.DataFrame(data)
    print(df.to_string(index=False))
    print('='*80)

def print_hyperparameter_table(mu_results, rho_results):
    """Print hyperparameter sensitivity table."""
    print('\n' + '='*60)
    print('HYPERPARAMETER SENSITIVITY ANALYSIS')
    print('='*60)

    print('\nFedProx (µ parameter):')
    print('-' * 40)
    fedprox_data = []
    for mu_val in sorted(mu_results.keys()):
        final_acc = mu_results[mu_val]['test_acc'][-1]
        fedprox_data.append({
            'µ Value': f'{mu_val:.3f}',
            'Final Test Acc (%)': f'{final_acc:.2f}'
        })
    df_fedprox = pd.DataFrame(fedprox_data)
    print(df_fedprox.to_string(index=False))

    print('\n\nFedSAM (ρ parameter):')
    print('-' * 40)
    fedsam_data = []
    for rho_val in sorted(rho_results.keys()):
        final_acc = rho_results[rho_val]['test_acc'][-1]
        fedsam_data.append({
            'ρ Value': f'{rho_val:.3f}',
            'Final Test Acc (%)': f'{final_acc:.2f}'
        })
    df_fedsam = pd.DataFrame(fedsam_data)
    print(df_fedsam.to_string(index=False))
    print('='*60)

# --------------------------------------------------------------
# 12. TABLE PRINTING FUNCTIONS
# --------------------------------------------------------------
def print_summary_table(all_results, methods_config):
    """Print a comprehensive summary table."""
    print('\n' + '='*120)
    print('COMPREHENSIVE RESULTS SUMMARY TABLE')
    print('='*120)

    # Prepare data for table
    data = []
    label_map = {'fedavg': 'FedAvg', 'fedprox': 'FedProx', 'scaffold': 'SCAFFOLD',
                 'fedgh': 'FedGH', 'fedsam': 'FedSAM'}

    communication_costs = {
        'fedavg': 1.0, 'fedprox': 1.0, 'scaffold': 2.0, 'fedgh': 1.0, 'fedsam': 1.0
    }
    computation_costs = {
        'fedavg': 1.0, 'fedprox': 1.05, 'scaffold': 1.1, 'fedgh': 1.02, 'fedsam': 2.0
    }

    for method_name in methods_config.keys():
        final_test_acc = all_results[method_name]['test_acc'][-1]
        final_test_loss = all_results[method_name]['test_loss'][-1]
        final_train_acc = all_results[method_name]['train_acc'][-1]
        avg_divergence = np.mean(all_results[method_name]['divergence'])
        gen_gap = final_train_acc - final_test_acc

        data.append({
            'Method': label_map[method_name],
            'Test Acc (%)': f'{final_test_acc:.2f}',
            'Test Loss': f'{final_test_loss:.4f}',
            'Train Acc (%)': f'{final_train_acc:.2f}',
            'Gen Gap (%)': f'{gen_gap:.2f}',
            'Avg Divergence': f'{avg_divergence:.5f}',
            'Comm Cost': f'{communication_costs[method_name]:.1f}x',
            'Comp Cost': f'{computation_costs[method_name]:.1f}x'
        })

    df = pd.DataFrame(data)
    print(df.to_string(index=False))
    print('='*120)

def print_convergence_table(all_results, methods_config, threshold=30.0):
    """Print convergence analysis table."""
    print('\n' + '='*80)
    print(f'CONVERGENCE ANALYSIS (Threshold: {threshold}% Test Accuracy)')
    print('='*80)

    data = []
    label_map = {'fedavg': 'FedAvg', 'fedprox': 'FedProx', 'scaffold': 'SCAFFOLD',
                 'fedgh': 'FedGH', 'fedsam': 'FedSAM'}

    for method_name in methods_config.keys():
        test_accs = all_results[method_name]['test_acc']

        # Find convergence round
        converged_round = None
        for i, acc in enumerate(test_accs):
            if acc >= threshold:
                converged_round = i + 1
                break

        # Calculate improvement over rounds
        if len(test_accs) >= 10:
            early_acc = np.mean(test_accs[:10])
            late_acc = np.mean(test_accs[-10:])
            improvement = late_acc - early_acc
        else:
            improvement = test_accs[-1] - test_accs[0]

        data.append({
            'Method': label_map[method_name],
            'Converged at Round': converged_round if converged_round else 'No',
            'Early Acc (R1-10)': f'{early_acc:.2f}%' if len(test_accs) >= 10 else f'{test_accs[0]:.2f}%',
            'Late Acc (R41-50)': f'{late_acc:.2f}%' if len(test_accs) >= 10 else f'{test_accs[-1]:.2f}%',
            'Improvement': f'{improvement:.2f}%'
        })

    df = pd.DataFrame(data)
    print(df.to_string(index=False))
    print('='*80)

def print_hyperparameter_table(mu_results, rho_results):
    """Print hyperparameter sensitivity table."""
    print('\n' + '='*60)
    print('HYPERPARAMETER SENSITIVITY ANALYSIS')
    print('='*60)

    print('\nFedProx (µ parameter):')
    print('-' * 40)
    fedprox_data = []
    for mu_val in sorted(mu_results.keys()):
        final_acc = mu_results[mu_val]['test_acc'][-1]
        fedprox_data.append({
            'µ Value': f'{mu_val:.3f}',
            'Final Test Acc (%)': f'{final_acc:.2f}'
        })
    df_fedprox = pd.DataFrame(fedprox_data)
    print(df_fedprox.to_string(index=False))

    print('\n\nFedSAM (ρ parameter):')
    print('-' * 40)
    fedsam_data = []
    for rho_val in sorted(rho_results.keys()):
        final_acc = rho_results[rho_val]['test_acc'][-1]
        fedsam_data.append({
            'ρ Value': f'{rho_val:.3f}',
            'Final Test Acc (%)': f'{final_acc:.2f}'
        })
    df_fedsam = pd.DataFrame(fedsam_data)
    print(df_fedsam.to_string(index=False))
    print('='*60)
```
